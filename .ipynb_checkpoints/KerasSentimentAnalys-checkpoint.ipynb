```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.datasets import imdb
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
import matplotlib.pyplot as plt
```

```python
# 1. Load dataset (giới hạn 10000 từ phổ biến nhất)
vocab_size = 10000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)
```

```python
# 2. Tiền xử lý dữ liệu
max_len = 200  # Giới hạn độ dài của mỗi review
x_train = pad_sequences(x_train, maxlen=max_len)
x_test = pad_sequences(x_test, maxlen=max_len)
```

```python
# 3. Xây dựng mô hình
model = Sequential()
model.add(Embedding(vocab_size, 32))
model.add(LSTM(32))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 4. Compile mô hình
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 5. Huấn luyện mô hình
history = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_test, y_test)) # Gán kết quả huấn luyện vào biến history
```

```python
# Vẽ biểu đồ Loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
```

```python
# Vẽ biểu đồ Accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='lower right')
plt.show()
```

```python
# 6. Đánh giá mô hình
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Loss: {loss:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```

```python
# 7. Hàm dự đoán sentiment cho một review mới
def predict_sentiment(review, word_index=imdb.get_word_index()):
    # Chuyển review thành chuỗi số
    review_sequence = []
    for word in review.split():
        index = word_index.get(word)
        if index is not None and index < vocab_size:
            review_sequence.append(index + 3) # +3 vì 0, 1, 2 là reserved indices

    review_sequence = pad_sequences([review_sequence], maxlen=max_len)
    prediction = model.predict(review_sequence)[0][0]
    return "Positive" if prediction > 0.5 else "Negative"
```

```python
# Ví dụ sử dụng
new_review = "This movie is amazing! I loved it."
sentiment = predict_sentiment(new_review)
print(f"Sentiment of '{new_review}': {sentiment}")

new_review = "This is the worst movie I've ever seen. Terrible acting and plot."
sentiment = predict_sentiment(new_review)
print(f"Sentiment of '{new_review}': {sentiment}")
```
